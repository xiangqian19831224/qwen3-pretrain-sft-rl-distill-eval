# Chat Template

```python
# Define structured output format for mathematical reasoning
REASONING_START = '<THINK>'  # Begin reasoning section
REASONING_END = '</THINK>'  # End reasoning section
ANSWER_START = '<ANSWER>'  # Begin final answer
ANSWER_END = '</ANSWER>'  # End final answer

# System prompt adapted for medical reasoning (inspired by paper's complex reasoning emphasis)
'''
ä½ æ˜¯ä¸€ååŒ»å­¦æ¨ç†åŠ©æ‰‹ã€‚å½“ç»™å‡ºä¸€ä¸ªåŒ»å­¦é—®é¢˜æ—¶ï¼š
åœ¨ {REASONING_START} å’Œ {REASONING_END} ä¹‹é—´å±•ç¤ºä½ é€æ­¥çš„å¤æ‚æ¨ç†è¿‡ç¨‹ï¼ˆåŒ…æ‹¬åæ€ã€å›æº¯ä»¥åŠæ›¿ä»£è·¯å¾„ï¼‰ã€‚
åœ¨ {ANSWER_START} å’Œ {ANSWER_END} ä¹‹é—´ç»™å‡ºä½ çš„æœ€ç»ˆç­”æ¡ˆã€‚
è¡¨è¾¾è¦ç²¾ç¡®ï¼Œéœ€è€ƒè™‘åŒ»å­¦ä¸­çš„åˆ«å/åŒä¹‰è¯ï¼Œå¹¶æ¸…æ™°å±•ç¤ºæ‰€æœ‰æ¨ç†ä¸æ€è€ƒæ­¥éª¤ã€‚
'''
SYSTEM_PROMPT = f'''You are a medical reasoning assistant. When given a medical problem:
1. Show your step-by-step complex reasoning (including reflection, backtracking, and alternative paths) between {REASONING_START} and {REASONING_END}.
2. Provide your final answer between {ANSWER_START} and {ANSWER_END}.
3. Be precise, consider medical aliases/synonyms, and show all deliberation steps clearly.'''

print(SYSTEM_PROMPT)
```

```python
'''
æ¨¡æ¿è§£æä¸¾ä¾‹:
1.æ¶ˆæ¯
    messages = [
          {"role": "system", "content": "You are a medical reasoning assistant."},
          {"role": "user", "content": "ä»€ä¹ˆæ˜¯ACSï¼Ÿ"},
          {"role": "assistant", "content": "ACS æŒ‡æ€¥æ€§å† è„‰ç»¼åˆå¾"},
          {"role": "user", "content": "å¦‚ä½•è¯Šæ–­ï¼Ÿ"},
        ]
2.æ¸²æŸ“ç»“æœï¼š
    You are a medical reasoning assistant.</s>
    ä»€ä¹ˆæ˜¯ACSï¼Ÿ
    ACS æŒ‡æ€¥æ€§å† è„‰ç»¼åˆå¾</s>
    å¦‚ä½•è¯Šæ–­ï¼Ÿ
    <REASONING_START>
'''

chat_template = (  # Build and assign chat_template to the tokenizer
    # If the very first message is a SYSTEM role, print it + <eos>:
    "{% if messages[0]['role'] == 'system' %}"
    "{{ messages[0]['content'] + eos_token }}"
    "{% set loop_messages = messages[1:] %}"
    "{% else %}"
    # Otherwise, inject our system_prompt + <eos>:
    "{{ '{system_prompt}' + eos_token }}"
    "{% set loop_messages = messages %}"
    "{% endif %}"

    # Now loop over the remaining messages (either user or assistant):
    "{% for message in loop_messages %}"
    "{% if message['role'] == 'user' %}"
    "{{ message['content'] }}"
    "{% elif message['role'] == 'assistant' %}"
    "{{ message['content'] + eos_token }}"
    "{% endif %}"
    "{% endfor %}"

    # If we asked for "add_generation_prompt", append <REASONING> to the end:
    "{% if add_generation_prompt %}{{ '{reasoning_start}' }}"
    "{% endif %}"
)
# Replace with out specific template:
tokenizer.chat_template = chat_template \
    .replace("'{system_prompt}'", f"'{SYSTEM_PROMPT}'") \
    .replace("'{reasoning_start}'", f"'{REASONING_START}'")
```

```python
example_messages = [  # Quick sanity check of the template
    {'role': 'user', 'content': 'What is the most severe complication of dengue fever?'},
    {'role': 'assistant', 'content': (
        f'{REASONING_START}'
        'Dengue fever can lead to severe forms like dengue hemorrhagic fever. '
        'Considering symptoms and progression, the most severe is plasma leakage leading to shock.'
        f'{REASONING_END}{ANSWER_START}Dengue shock syndrome{ANSWER_END}'
    )},
    {'role': 'user', 'content': 'What drug is used for hypertension?'},
]
print(tokenizer.apply_chat_template(example_messages, tokenize=False, add_generation_prompt=True))
```

# Pre Fine-tuning (SFT)

## Data preparation

```python
def format_dataset(x):
    # Format the dataset to follow our GRPO style formatting (adapted for medical SFT data)

    # The medical-o1-reasoning-SFT has 'Complex CoT' and 'Response' fields
    # 'Response' is the final answer, 'question' is the problem
    expected_answer, problem = x['Response'], x['Question']
    # Use the provided Complex CoT as reasoning
    thoughts = x['Complex_CoT'].strip()
    # Custom formatting
    final_prompt = REASONING_START + thoughts + REASONING_END + ANSWER_START + expected_answer + ANSWER_END

    return [
        {'role': 'system', 'content': SYSTEM_PROMPT},
        {'role': 'user', 'content': problem},
        {'role': 'assistant', 'content': final_prompt},
    ]


'''
FreedomIntelligence/medical-o1-reasoning-SFT
1.ç®€ä»‹
    è¿™ä¸ªæ•°æ®é›†ç”¨äºå¾®è°ƒ HuatuoGPT-o1ï¼Œè¿™æ˜¯ä¸€ä¸ªé¢å‘é«˜çº§åŒ»å­¦æ¨ç†è®¾è®¡çš„åŒ»å­¦å¤§è¯­è¨€æ¨¡å‹ã€‚
    è¯¥æ•°æ®é›†ç”± GPT-4o æ„å»ºï¼ŒGPT-4o ä¼šé’ˆå¯¹å¯éªŒè¯çš„åŒ»å­¦é—®é¢˜æœç´¢è§£å†³æ–¹æ¡ˆï¼Œå¹¶é€šè¿‡**åŒ»å­¦éªŒè¯æ¨¡å‹ï¼ˆmedical verifierï¼‰**å¯¹ç»“æœè¿›è¡Œæ ¡éªŒã€‚
2.å­—æ®µ
    questionï¼š   é—®é¢˜
    responseï¼š   å›ç­”
    Complex_COT: æ¨ç†
'''
sft_dataset = load_dataset('FreedomIntelligence/medical-o1-reasoning-SFT', 'en', split='train').to_pandas()
sft_dataset['messages'] = sft_dataset.apply(format_dataset, axis=1)
print(tokenizer.apply_chat_template(sft_dataset['messages'][0], tokenize=False))
sft_dataset
```

```python
from datasets import Dataset, DatasetDict
import pandas as pd

# Truncate pre fine-tuning sft_dataset to max_seq_length / 2 since we don't want too long reasoning traces
sft_dataset['seq_length'] = sft_dataset['messages'].apply(lambda x: len(tokenizer.apply_chat_template(x)))
print('Token-length percentiles (50/90/99):', np.percentile(sft_dataset['seq_length'], [50, 90, 99]))

threshold = max_seq_length / 2
sft_dataset_filtered = sft_dataset.loc[sft_dataset['seq_length'] <= threshold].copy()
print(f'Remaining for training (<= {threshold} tokens): {len(sft_dataset_filtered)}/{len(sft_dataset)}')

sft_dataset_filtered['text'] = tokenizer.apply_chat_template(sft_dataset_filtered['messages'].values.tolist(),
                                                             tokenize=False)
sft_dataset_filtered = Dataset.from_pandas(sft_dataset_filtered)
sft_dataset_filtered
```

## Multi-reward design

```python
# Match the reasoning sections and answers
match_format = re.compile(
    rf'{REASONING_END}.*?'  # We always prepend REASONING_START
    rf'{ANSWER_START}(.+?){ANSWER_END}'  # Answer section with capture group (text, not number)
    rf'[\s]{{0,}}(?:{re.escape(tokenizer.eos_token)})?'  # Add optional EOS token matching
    rf'[\s]{{0,}}$',  # Optional whitespace at end
    flags=re.MULTILINE | re.DOTALL,  # Multi-line matching with . matching newlines
)


# è®¡ç®—ä¸æ­£ç¡®ç­”æ¡ˆçš„ç›¸ä¼¼åº¦
def verify(guess, true_answer):
    verifier_input = VERIFIER_TEMPLATE.format(guess.strip(), true_answer.strip(), verifier_tokenizer.eos_token)
    input_batch = verifier_tokenizer([verifier_input], return_tensors='pt').to(verifier_model.device)
    with torch.no_grad():
        logits = verifier_model(**input_batch,
                                return_dict=True).logits  # logits: [batch_size, num_labels] ä¾‹å¦‚ï¼š [[ 1.2, 3.4 ]]
    probabilities = F.softmax(logits, dim=-1)  # probabilities: [batch_size, num_labels] ä¾‹å¦‚ï¼š [[0.10, 0.90]]
    return probabilities[0, 1].item()  # 0.90
```

```python
def match_format_strictly(completions, **kwargs) -> list[float]:
    ''' Reward Function 1: Exact Format Compliance
    High reward (3.0) for perfect format adherence. Ensure model learns the complete structured output pattern
    '''
    return [
        3.0 if match_format.search(completion[0]['content']) else 0.0
        for completion in completions
    ]
```

```python
# If it fails, reward the model if it at least follows the format partially, by counting each symbol
def match_format_softly(completions, **kwargs) -> list[float]:
    ''' Reward Function 2: Partial Format Credit
    Graduated scoring for format elements. Encourage learning individual components even if not perfect
    '''
    rewards = []
    for completion in completions:
        reward = 0
        response = completion[0]['content']

        # Count how many keywords are seen - we penalize if too many!
        # Award +0.5 for correct token count, -0.5 for wrong count
        # reward += 0.5 if response.count(REASONING_START) == 1 else -0.5  # Prepended
        reward += 0.5 if response.count(REASONING_END) == 1 else -0.5
        reward += 0.5 if response.count(ANSWER_START) == 1 else -0.5
        reward += 0.5 if response.count(ANSWER_END) == 1 else -0.5
        rewards.append(reward)
    return rewards
```

```python
# Extract the generated answer, and reward or penalize it
def check_answer_correctness(completions, answer, **kwargs) -> list[float]:
    ''' Reward Function 3: Graduated scoring for medical accuracy using verifier
    - 5.0: Verifier confirms full alignment (True)
    - 2.0: Partial credit if verifier prob > 0.5 but not full (use prob for gradation)
    - -2.5: Wrong answer (False)
    Handles aliases via semantic verification
    '''
    responses = [completion[0]['content'] for completion in completions]
    extracted_responses = [  # Extract answers using format pattern
        guess.group(1) if (guess := match_format.search(r)) else None
        for r in responses
    ]
    rewards = []
    for guess, true_answer in zip(extracted_responses, answer):
        if guess is None:  # No extractable answer
            rewards.append(-2.0)
            continue

        prob_true = verify(guess, true_answer)  # Use verifier to check semantic alignment
        if prob_true > 0.9:
            rewards.append(5.0)  # High confidence correct
        elif prob_true > 0.7:
            rewards.append(3.5)  # Strong alignment
        elif prob_true > 0.5:
            rewards.append(2.0)  # Partial/approximate
        elif prob_true > 0.3:
            rewards.append(1.5)  # Reasonable attempt
        else:
            rewards.append(-2.5)  # Incorrect
    return rewards
```

```python
# é…ç½® GRPOï¼ˆGeneralized Reinforcement Policy Optimizationï¼‰è®­ç»ƒå‚æ•°ï¼Œé’ˆå¯¹æ•°å­¦æ¨ç†ä»»åŠ¡
training_args = GRPOConfig(
    output_dir=f'./output/tmp/{model_name}',  # æ£€æŸ¥ç‚¹ï¼ˆcheckpointï¼‰å’Œæ—¥å¿—ä¿å­˜ç›®å½•
    vllm_sampling_params=vllm_sampling_params,  # vLLM çš„é‡‡æ ·å‚æ•°ï¼Œç”¨äºæ§åˆ¶ç”Ÿæˆç­–ç•¥
    # è®­ç»ƒé€Ÿåº¦æ§åˆ¶
    num_train_epochs=1,  # æ€»è®­ç»ƒè½®æ•°ï¼ˆepochï¼‰ï¼Œ1 è¡¨ç¤ºéå†è®­ç»ƒæ•°æ®é›†ä¸€æ¬¡
    per_device_train_batch_size=1,  # æ¯ä¸ª GPU/è®¾å¤‡ä¸Šçš„è®­ç»ƒ batch å¤§å°
    gradient_accumulation_steps=16,  # æ¢¯åº¦ç´¯ç§¯æ­¥æ•°ï¼Œæœ‰æ•ˆ batch = 4 * 16 = 64ï¼Œå¯ç¼“è§£æ˜¾å­˜å‹åŠ›
    # ç²¾åº¦ä¸ä¼˜åŒ–
    scale_rewards='batch',  # å¥–åŠ±å½’ä¸€åŒ–æ–¹å¼ï¼šåœ¨ batch å†…è¿›è¡Œå‡å€¼å’Œæ ‡å‡†å·®å½’ä¸€åŒ–ï¼Œæœ‰åŠ©äºç¨³å®šè®­ç»ƒ
    loss_type='dr_grpo',  # æŸå¤±ç±»å‹ï¼Œdr_grpo å¯å®Œå…¨å»é™¤å›ç­”é•¿åº¦åå·®ï¼Œç”¨å¸¸æ•°é™¤è€Œéåºåˆ—é•¿åº¦
    optim='adamw_8bit',  # ä¼˜åŒ–å™¨ç±»å‹ï¼Œ8bit AdamWï¼ŒèŠ‚çœæ˜¾å­˜
    weight_decay=0.1,  # æƒé‡è¡°å‡ï¼ˆL2 æ­£åˆ™åŒ–ï¼‰ï¼Œé˜²æ­¢è¿‡æ‹Ÿåˆ
    max_grad_norm=0.1,  # æ¢¯åº¦è£å‰ªé˜ˆå€¼ï¼Œé˜²æ­¢æ¢¯åº¦çˆ†ç‚¸ï¼Œæé«˜è®­ç»ƒç¨³å®šæ€§
    bf16=is_bfloat16_supported(),  # æ˜¯å¦å¯ç”¨ bf16 æ··åˆç²¾åº¦è®­ç»ƒï¼ŒåŠ é€Ÿè®­ç»ƒå¹¶é™ä½æ˜¾å­˜æ¶ˆè€—
    learning_rate=1e-5,  # å­¦ä¹ ç‡ï¼Œè¾ƒå°å€¼å¯é˜²æ­¢æ¨ç†èƒ½åŠ›è¢«ç ´å
    lr_scheduler_type='cosine',  # å­¦ä¹ ç‡è°ƒåº¦ç­–ç•¥ï¼Œcosine æ›²çº¿é€æ¸é™ä½å­¦ä¹ ç‡
    warmup_ratio=0.05,  # å­¦ä¹ ç‡é¢„çƒ­æ¯”ä¾‹ï¼Œå‰ 5% æ­¥æ•°é€æ¸å‡è‡³ç›®æ ‡å­¦ä¹ ç‡
    # æ–‡æœ¬ç”Ÿæˆæ§åˆ¶
    temperature=1.0,  # ç”Ÿæˆæ¸©åº¦ï¼Œæ§åˆ¶é‡‡æ ·éšæœºæ€§ï¼Œ1.0 è¡¨ç¤ºè‡ªç„¶æ¢ç´¢
    num_generations=4,  # æ¯æ¡è¾“å…¥ prompt ç”Ÿæˆçš„æ ·æœ¬æ•°é‡
    max_prompt_length=max_prompt_length,  # æœ€å¤§ prompt é•¿åº¦ï¼Œè¶…å‡ºéƒ¨åˆ†æˆªæ–­
    max_completion_length=max_completion_length,  # æœ€å¤§ç”Ÿæˆé•¿åº¦ï¼Œè¶…è¿‡åˆ™æˆªæ–­
    # æ—¥å¿—è®°å½•ä¸æ¨¡å‹ä¿å­˜
    report_to='none',  # æ˜¯å¦æŠ¥å‘Šè®­ç»ƒæŒ‡æ ‡åˆ°å¤–éƒ¨å¹³å°ï¼Œå¦‚ WandBï¼Œè¿™é‡Œç¦ç”¨
    logging_steps=79,  # æ¯éš”å¤šå°‘æ­¥è®°å½•ä¸€æ¬¡è®­ç»ƒæ—¥å¿—
    logging_strategy='steps',  # æ—¥å¿—è®°å½•ç­–ç•¥ï¼ŒæŒ‰æ­¥ï¼ˆstepï¼‰è®°å½•
    save_total_limit=1,  # æœ€å¤šä¿ç•™å¤šå°‘ä¸ª checkpointï¼Œæ—§çš„ä¼šè¢«åˆ é™¤
    # max_steps=100,  # å¯é€‰å‚æ•°ï¼šæ€»è®­ç»ƒæ­¥æ•°ï¼Œæœªè®¾ç½®åˆ™æ ¹æ® epoch å’Œæ•°æ®é›†å¤§å°è®¡ç®—
)
```

## Train the model

```python
%%time

'''
train_dataset: æ•°æ®æ ¼å¼
{
    'prompt': [  # Create conversation with system prompt for structured reasoning
        {'role': 'system', 'content': SYSTEM_PROMPT},
        {'role': 'user', 'content': example['Open-ended Verifiable Question']},
    ],
    # Ground truth answer for reward functions (text, not numerical)
    'answer': example['Ground-True Answer'].strip()
}
'''
trainer = GRPOTrainer(  # Initialize GRPO trainer with multi-reward system (improvement over paper's single-reward PPO)
    model=model,  # LoRA-adapted quantized model
    processing_class=tokenizer,
    train_dataset=train_dataset,  # Processed medical dataset
    args=training_args,  # Training configuration
    reward_funcs=[  # 3 complementary reward functions
        match_format_strictly,  # Perfect structure compliance
        match_format_softly,  # Partial format credit
        check_answer_correctness,  # Semantic accuracy via verifier (handles aliases)
    ]
)
trainer.train()
trainer.save_model(f'./output/{model_name}_grpo')
```

```python
# https://docs.unsloth.ai/models/qwen3-how-to-run-and-fine-tune#official-recommended-settings

# åˆ›å»º vLLM çš„é‡‡æ ·å‚æ•°å¯¹è±¡ï¼ˆSamplingParamsï¼‰
# è¯¥å¯¹è±¡ç›´æ¥å½±å“ decoder åœ¨æ¯ä¸€æ­¥ token ç”Ÿæˆæ—¶çš„å€™é€‰ç©ºé—´ä¸é‡‡æ ·ç­–ç•¥
# æœ¬è´¨ä¸Šæ˜¯åœ¨æ§åˆ¶ã€Œæ¢ç´¢ï¼ˆdiversityï¼‰ã€ä¸ã€Œç¡®å®šæ€§ï¼ˆstabilityï¼‰ã€ä¹‹é—´çš„æƒè¡¡
sampling_params = SamplingParams(
    temperature=0.6,
    # æ¸©åº¦ç³»æ•°ï¼ˆTemperature Scalingï¼‰ï¼š
    # - ä¼šå¯¹ logits åš softmax ä¹‹å‰çš„ç¼©æ”¾ï¼ˆlogits / temperatureï¼‰
    # - temperature < 1ï¼šæ‹‰å¤§é«˜æ¦‚ç‡ token ä¸ä½æ¦‚ç‡ token çš„å·®è·ï¼Œè¾“å‡ºæ›´ç¨³å®šã€æ›´åå‘â€œæœ€ä¼˜è§£â€
    # - temperature > 1ï¼šå‹å¹³åˆ†å¸ƒï¼Œå¢åŠ éšæœºæ€§ä¸å‘æ•£åº¦
    # - 0.6 å¸¸ç”¨äºï¼šæ¨ç† / é—®ç­” / åŒ»ç–— / ä»£ç  ç­‰ã€Œè´¨é‡ä¼˜å…ˆã€åœºæ™¯
    # - ä¸ top_p / top_k è”åˆä½¿ç”¨æ—¶ï¼Œæ¸©åº¦ä¸»è¦å†³å®šâ€œæ¦‚ç‡åˆ†å¸ƒå½¢çŠ¶â€

    min_p=0.0,
    # æœ€å°æ¦‚ç‡é˜ˆå€¼ï¼ˆMinimum Probability Filteringï¼‰ï¼š
    # - åœ¨ softmax åï¼Œç›´æ¥ä¸¢å¼ƒæ¦‚ç‡ < min_p çš„ token
    # - ç±»ä¼¼äºä¸€ä¸ªâ€œç¡¬é˜ˆå€¼â€çš„ä½è´¨é‡ token è£å‰ª
    # - 0.0 è¡¨ç¤ºä¸å¯ç”¨è¯¥ç­–ç•¥ï¼ˆä¸åšæ¦‚ç‡ä¸‹é™è£å‰ªï¼‰
    # - åœ¨æŸäº› hallucination æ•æ„Ÿä»»åŠ¡ä¸­ï¼ˆå¦‚åŒ»ç–—/æ³•å¾‹ï¼‰ï¼Œå¯é€‚å½“è®¾ä¸º 0.01~0.05
    # - ä¸ top_p ä¸åŒï¼šmin_p æ˜¯â€œç»å¯¹æ¦‚ç‡â€ï¼Œtop_p æ˜¯â€œç´¯è®¡æ¦‚ç‡â€

    top_p=0.95,
    # æ ¸é‡‡æ ·ï¼ˆNucleus Samplingï¼‰ï¼š
    # - ä»æŒ‰æ¦‚ç‡æ’åºåçš„ token ä¸­ï¼Œé€‰æ‹©ç´¯è®¡æ¦‚ç‡ â‰¥ top_p çš„æœ€å°é›†åˆ
    # - åªåœ¨è¿™ä¸ªâ€œé«˜ç½®ä¿¡åº¦æ ¸å¿ƒé›†åˆâ€ä¸­è¿›è¡Œé‡‡æ ·
    # - èƒ½æœ‰æ•ˆæˆªæ–­é•¿å°¾ä½æ¦‚ç‡ tokenï¼Œå‡å°‘èƒ¡ç¼–ä¹±é€ 
    # - 0.9~0.95 æ˜¯å·¥ä¸šç•Œæœ€å¸¸è§åŒºé—´
    # - ä¸ temperature è”åˆï¼štemperature å†³å®šåˆ†å¸ƒå½¢çŠ¶ï¼Œtop_p å†³å®šâ€œä¿ç•™å¤šå¤§æ ¸å¿ƒâ€

    top_k=20,
    # Top-K é‡‡æ ·ï¼š
    # - ä»…ä¿ç•™æ¦‚ç‡æœ€é«˜çš„ K ä¸ª tokenï¼Œå…¶ä½™ç›´æ¥ä¸¢å¼ƒ
    # - æ˜¯ä¸€ç§â€œç¡¬æˆªæ–­â€çš„å€™é€‰ç©ºé—´é™åˆ¶
    # - K=20 è¡¨ç¤º decoder æ¯ä¸€æ­¥æœ€å¤šåªåœ¨ 20 ä¸ª token ä¸­é‡‡æ ·
    # - é€šå¸¸ä¸ top_p åŒæ—¶ä½¿ç”¨ï¼š
    #   - top_k æ§åˆ¶â€œä¸Šé™â€
    #   - top_p æ§åˆ¶â€œç´¯è®¡æ¦‚ç‡ä¸‹é™â€
    # - å¯¹æ¨ç†é€Ÿåº¦å’Œç¨³å®šæ€§éƒ½æœ‰å¸®åŠ©ï¼Œå°¤å…¶åœ¨å¤§è¯è¡¨æ¨¡å‹ä¸­

    max_tokens=max_completion_length,
    # æœ€å¤§ç”Ÿæˆ token æ•°ï¼š
    # - é™åˆ¶æ¨¡å‹åœ¨ä¸€æ¬¡è¯·æ±‚ä¸­æœ€å¤šç”Ÿæˆå¤šå°‘ token
    # - é˜²æ­¢æ¨¡å‹æ— é™ç»­å†™ï¼ˆå°¤å…¶åœ¨æ—  stop token æ—¶ï¼‰
    # - åœ¨è¯„æµ‹ / æ‰¹é‡æ¨ç†ä¸­ï¼Œæ˜¯æ˜¾å­˜ã€æ—¶å»¶ã€ååçš„é‡è¦æ§åˆ¶å‚æ•°
    # - é€šå¸¸ä¸æ•°æ®é›†ä»»åŠ¡ï¼ˆå¦‚ QA / å¤šé€‰ / è§£é‡Šå‹å›ç­”ï¼‰å¼ºç›¸å…³

    truncate_prompt_tokens=threshold,
    # Prompt æˆªæ–­é•¿åº¦ï¼š
    # - å½“è¾“å…¥ prompt è¶…è¿‡è¯¥é•¿åº¦æ—¶ï¼Œåªä¿ç•™æœ€åçš„ threshold ä¸ª token
    # - ç”¨äºé˜²æ­¢è¶…é•¿ä¸Šä¸‹æ–‡å¯¼è‡´ï¼š
    #   1ï¼‰æ˜¾å­˜å ç”¨è¿‡å¤§
    #   2ï¼‰attention è®¡ç®—å˜æ…¢
    #   3ï¼‰KV cache çˆ†ç‚¸
    # - å¯¹å¤šè½®å¯¹è¯ / RAG åœºæ™¯å°¤ä¸ºé‡è¦
    # - ä¸€èˆ¬ä¿ç•™â€œæœ€è¿‘ä¸Šä¸‹æ–‡â€æ¯”â€œæœ€æ—©ä¸Šä¸‹æ–‡â€æ›´æœ‰ä»·å€¼
)


# ç¼–è¯‘æ­£åˆ™è¡¨è¾¾å¼æ¨¡å¼ï¼Œç”¨äºä»æ¨¡å‹ç”Ÿæˆçš„å›ç­”ä¸­æå–ç­”æ¡ˆå­—æ¯
match_answer_letter = re.compile(
    # æ„å»ºæ­£åˆ™æ¨¡å¼ï¼šæŸ¥æ‰¾ANSWER_STARTæ ‡è®°åçš„[A-D]å­—æ¯
    rf'{ANSWER_START}.*?[\s]{{0,}}([A-D])',

    # æ ‡å¿—ä½ï¼š
    flags=re.MULTILINE | re.DOTALL

    # æ¨¡å¼è§£é‡Šï¼š
    # {ANSWER_START} - åŒ¹é…ç­”æ¡ˆå¼€å§‹æ ‡è®°ï¼ˆå¦‚<ANSWER>ï¼‰
    # .*? - éè´ªå©ªåŒ¹é…ä»»æ„å­—ç¬¦ï¼ˆé™¤æ¢è¡Œç¬¦å¤–ï¼‰
    # [\s]{{0,}} - åŒ¹é…0åˆ°å¤šä¸ªç©ºç™½å­—ç¬¦ï¼ˆç©ºæ ¼ã€åˆ¶è¡¨ç¬¦ç­‰ï¼‰
    # ([A-D]) - æ•è·ç»„ï¼šåŒ¹é…å¹¶æ•è·Aã€Bã€Cã€Dä¸­çš„ä»»æ„ä¸€ä¸ªå­—æ¯
    #         è¿™æ˜¯æˆ‘ä»¬è¦æå–çš„æœ€ç»ˆç­”æ¡ˆé€‰é¡¹

    # re.MULTILINE - å¤šè¡Œæ¨¡å¼ï¼š^å’Œ$åŒ¹é…æ¯è¡Œçš„å¼€å§‹å’Œç»“æŸ
    # re.DOTALL - ç‚¹åŒ¹é…æ¢è¡Œï¼šä½¿.ä¹Ÿèƒ½åŒ¹é…æ¢è¡Œç¬¦ï¼Œå¤„ç†è·¨è¡Œçš„ç­”æ¡ˆ
)
```

```python
# å®šä¹‰ä¸€ä¸ªç¤ºä¾‹æ–‡æœ¬ï¼Œè¯¢é—®é«˜è¡€å‹æ²»ç–—è¯ç‰©
example_text = 'What drug is used for hypertension?'

# ä½¿ç”¨æ¨¡å‹çš„å¿«é€Ÿç”ŸæˆåŠŸèƒ½è¿›è¡Œæ¨ç†ï¼Œè¿™ä¸ªè°ƒç”¨æ˜¯åœ¨GRPOè®­ç»ƒä¹‹å‰æµ‹è¯•æ¨¡å‹çš„åŸå§‹èƒ½åŠ›
# model.fast_generate() - ä½¿ç”¨vLLMä¼˜åŒ–çš„å¿«é€Ÿç”Ÿæˆæ–¹æ³•
# å‚æ•°1: example_text - è¾“å…¥çš„é—®é¢˜æ–‡æœ¬
# sampling_params=sampling_params - é‡‡æ ·å‚æ•°ï¼Œæ§åˆ¶ç”Ÿæˆæ–‡æœ¬çš„éšæœºæ€§å’Œè´¨é‡
# lora_request=None - ä¸ä½¿ç”¨LoRAé€‚é…å™¨ï¼Œç›´æ¥ä½¿ç”¨åŸºç¡€æ¨¡å‹
# [0] - è·å–ç¬¬ä¸€ä¸ªè¯·æ±‚çš„ç»“æœï¼ˆå¯èƒ½æ”¯æŒæ‰¹é‡å¤„ç†ï¼‰
# .outputs[0] - è·å–ç¬¬ä¸€ä¸ªè¾“å‡ºç»“æœ
# .text - è·å–ç”Ÿæˆçš„æ–‡æœ¬å†…å®¹
# print() - å°†ç”Ÿæˆçš„å“åº”æ–‡æœ¬æ‰“å°åˆ°æ§åˆ¶å°
print(model.fast_generate(
    example_text,
    sampling_params=sampling_params,
    lora_request=None
)[0].outputs[0].text)
```

```python
# åˆ›å»ºä¸€ä¸ªç©ºå­—å…¸ï¼Œç”¨äºå­˜å‚¨ä»æ–‡ä»¶ä¸­åŠ è½½çš„å¼ é‡
tensors = {}

# ä½¿ç”¨å®‰å…¨å¼ é‡æ ¼å¼æ‰“å¼€è®­ç»ƒåçš„LoRAé€‚é…å™¨æƒé‡æ–‡ä»¶
# framework='pt' è¡¨ç¤ºä½¿ç”¨PyTorchæ¡†æ¶
with safe_open(f'./output/{model_name}_grpo/adapter_model.safetensors', framework='pt') as f:
    # éå†æ–‡ä»¶ä¸­çš„æ‰€æœ‰å¼ é‡é”®åï¼ˆæ¯ä¸ªé”®å¯¹åº”ä¸€ä¸ªæƒé‡çŸ©é˜µï¼‰
    for key in f.keys():
        # ä»æ–‡ä»¶ä¸­è·å–å¯¹åº”çš„å¼ é‡æ•°æ®
        tensor = f.get_tensor(key)

        # è®¡ç®—å¼ é‡ä¸­é›¶å€¼å…ƒç´ çš„æ¯”ä¾‹
        # (tensor == 0).sum() ç»Ÿè®¡é›¶å€¼å…ƒç´ æ•°é‡
        # tensor.numel() è·å–å¼ é‡æ€»å…ƒç´ æ•°é‡
        n_zeros = (tensor == 0).sum() / tensor.numel()

        # æ–­è¨€æ£€æŸ¥ï¼šç¡®ä¿å¼ é‡ä¸ä¼šå…¨éƒ¨ä¸ºé›¶
        # å¦‚æœå¼ é‡å…¨éƒ¨ä¸ºé›¶ï¼Œè¯´æ˜LoRAå‚æ•°æ²¡æœ‰æ›´æ–°ï¼Œè®­ç»ƒå¯èƒ½å¤±è´¥
        assert (n_zeros.item() != tensor.numel())
```

```python
# 1. å°†è¾“å…¥æ–‡æœ¬åº”ç”¨èŠå¤©æ¨¡æ¿ï¼ˆchat templateï¼‰æ ¼å¼åŒ–
#    - role: æŒ‡å®šè§’è‰²ï¼Œè¿™é‡Œæ˜¯ 'user'ï¼Œè¡¨ç¤ºç”¨æˆ·è¾“å…¥
#    - content: ç”¨æˆ·è¾“å…¥çš„å†…å®¹ï¼Œå³ example_text
#    - add_generation_prompt=True: åœ¨è¾“å…¥æœ«å°¾æ·»åŠ æ¨¡å‹ç”Ÿæˆçš„æç¤ºï¼Œä¾¿äºç”Ÿæˆå›ç­”
#    - tokenize=False: ä¸ç«‹å³å°†æ–‡æœ¬è½¬æ¢ä¸º tokenï¼Œä¿æŒåŸå§‹å­—ç¬¦ä¸²
text = tokenizer.apply_chat_template(
    [{'role': 'user', 'content': example_text}],
    add_generation_prompt=True,
    tokenize=False,
)

# 2. ä½¿ç”¨ fast_generate æ–¹æ³•ç”Ÿæˆæ¨¡å‹è¾“å‡º
#    - text: ä¸Šä¸€æ­¥ç”Ÿæˆçš„èŠå¤©æ¨¡æ¿æ–‡æœ¬
#    - sampling_params: é‡‡æ ·å‚æ•°ï¼ˆå¦‚ temperatureã€top_p ç­‰ï¼‰ï¼Œæ§åˆ¶ç”Ÿæˆæ–‡æœ¬çš„å¤šæ ·æ€§
#    - lora_request: åŠ è½½ LoRA æƒé‡ï¼Œç”¨äºå¯¹åŸæ¨¡å‹è¿›è¡Œè½»é‡å¾®è°ƒ
#        model.load_lora(f'./{model_name}_grpo') ä¼šåŠ è½½æŒ‡å®šè·¯å¾„ä¸‹çš„ LoRA æƒé‡
#    è¯´æ˜: å› ä¸ºæ˜¯æµ‹è¯• LoRAï¼Œæ‰€ä»¥ä¸ä½¿ç”¨ç³»ç»Ÿæç¤ºï¼ˆsystem promptï¼‰ï¼Œ
#         ç›®çš„æ˜¯å°½é‡å‡å°‘å¯¹åŸæ¨¡å‹æ¨ç†èƒ½åŠ›çš„å½±å“
print(model.fast_generate(
    text,
    sampling_params=sampling_params,
    lora_request=model.load_lora(f'./output/{model_name}_grpo'),
)[0].outputs[0].text)  # 3. è¾“å‡ºç”Ÿæˆçš„æ–‡æœ¬ç»“æœ
```

## Performance on benchmark datasets

```python
def format_mcq_prompt(example, dataset_name):
    """
    æ ¹æ®ä¸åŒçš„æ•°æ®é›†ï¼Œå°†å¤šé€‰é¢˜ï¼ˆMCQï¼‰ç¤ºä¾‹æ ¼å¼åŒ–ä¸ºé€‚åˆæ¨¡å‹è¾“å…¥çš„ promptã€‚
    å‚æ•°:
        example (dict): åŒ…å«é—®é¢˜ã€é€‰é¡¹åŠç­”æ¡ˆçš„ç¤ºä¾‹ã€‚
        dataset_name (str): æ•°æ®é›†åç§°ï¼Œå¯ä¸º 'medqa', 'medmcqa', 'pubmedqa'ã€‚
    è¿”å›:
        dict: åŒ…å« 'prompt' å’Œ 'answer' çš„å­—å…¸ï¼Œæ–¹ä¾¿åç»­ç”¨äºè®­ç»ƒæˆ–è¯„ä¼°æ¨¡å‹ã€‚
    """

    # 1. æ„å»ºç”¨æˆ·æé—®éƒ¨åˆ†ï¼Œæç¤ºæ¨¡å‹åªè¿”å›æ­£ç¡®é€‰é¡¹å­—æ¯ï¼Œä¸è¦å¸¦é¢å¤–æ–‡æœ¬
    user_content = 'Please answer the following multiple-choice question, ensuring your response only '
    user_content += 'contain the correct option letter with no extra text:\n' + example['question'] + '\n'

    # 2. é’ˆå¯¹ä¸åŒçš„æ•°æ®é›†ï¼Œå¤„ç†é€‰é¡¹å’Œç­”æ¡ˆ
    if dataset_name == 'medqa':
        # medqa çš„é€‰é¡¹å­˜å‚¨åœ¨å­—å…¸ example['options'] ä¸­ï¼Œkey ä¸ºé€‰é¡¹å­—æ¯ï¼Œvalue ä¸ºé€‰é¡¹å†…å®¹
        user_content += '\n'.join([f'{k}. {v}' for k, v in example['options'].items()])
        # gold_letter ä¸ºæ­£ç¡®ç­”æ¡ˆçš„å­—æ¯ç´¢å¼•
        gold_letter = example['answer_idx']

    elif dataset_name == 'medmcqa':
        # medmcqa çš„é€‰é¡¹åˆ†åˆ«å­˜å‚¨åœ¨ example['opa', 'opb', 'opc', 'opd']
        choices = [example['opa'], example['opb'], example['opc'], example['opd']]
        # æ„å»ºé€‰é¡¹å­—ç¬¦ä¸²ï¼ŒæŒ‰ A/B/C/D æ’åˆ—
        user_content += '\n'.join([f'{chr(65 + i)}. {choice}' for i, choice in enumerate(choices)])
        # gold_letter æ ¹æ®æ•´æ•°ç´¢å¼• cop è½¬æ¢ä¸ºå­—æ¯
        gold_letter = chr(ord('A') + example['cop'])

    elif dataset_name == 'pubmedqa':
        # pubmedqa éœ€è¦é™„åŠ ä¸Šä¸‹æ–‡ï¼ˆå¦‚æ‘˜è¦ä¿¡æ¯ï¼‰
        contexts = ' '.join(example['context']['contexts'])  # å°†å¤šä¸ªä¸Šä¸‹æ–‡æ®µè½æ‹¼æ¥ä¸ºå•ä¸ªå­—ç¬¦ä¸²
        # æ„å»ºæœ€ç»ˆ promptï¼ŒåŒ…æ‹¬ä¸Šä¸‹æ–‡ã€é—®é¢˜å’Œå›ºå®šé€‰é¡¹ A/B/C
        user_content = f'Context: {contexts}\n{user_content}A. yes\nB. no\nC. maybe'
        # gold_letter æ ¹æ® final_decision å­—æ®µæ˜ å°„ä¸º A/B/C
        gold_letter = 'A' if example['final_decision'] == 'yes' else 'B' if example['final_decision'] == 'no' else 'C'

    # 3. è¿”å›æ ¼å¼åŒ–åçš„ prompt å’Œæ­£ç¡®ç­”æ¡ˆ
    return {
        'prompt': [
            {'role': 'system', 'content': SYSTEM_PROMPT},  # ç³»ç»Ÿè§’è‰²æŒ‡ä»¤ï¼Œå®šä¹‰æ¨¡å‹è¡Œä¸º
            {'role': 'user', 'content': user_content},     # ç”¨æˆ·é—®é¢˜åŠé€‰é¡¹
        ],
        'answer': gold_letter.strip()  # æ­£ç¡®é€‰é¡¹å­—æ¯
    }
```

```python
# Load evaluation benchmarks as per paper: MedQA-USMLE (test), MedMCQA (validation), PubMedQA (PQA-L test equivalent)
# 1273 samples; gold in 'answer_idx' but use open-ended, verifier with options['answer_idx']
medqa_dataset = load_dataset('GBaker/MedQA-USMLE-4-options', split='test')
# 4183 samples; gold in 'cop' (choice index)
medmcqa_dataset = load_dataset('openlifescienceai/medmcqa',split='validation')
# 1000 questions in the PQA-L are used as the test set
pubmedqa_dataset = load_dataset('pubmed_qa', 'pqa_labeled',split='train')
```

```python
# æ§åˆ¶æ•°æ®é•¿åº¦



eval_datasets = {
    'MedQA_USLME_test': medqa_dataset.map(lambda x: format_mcq_prompt(x, 'medqa')),
    'MedMCQA_validation': medmcqa_dataset.map(lambda x: format_mcq_prompt(x, 'medmcqa')),
    'PubMedQA_test': pubmedqa_dataset.map(lambda x: format_mcq_prompt(x, 'pubmedqa')),
}
```

```python
%%time
# åˆ›å»ºä¸€ä¸ªç©ºå­—å…¸ï¼Œç”¨æ¥å­˜å‚¨æ¯ä¸ªæ•°æ®é›†çš„è¯„ä¼°ç»“æœ
results = {}

# éå†æ‰€æœ‰å¾…è¯„ä¼°çš„æ•°æ®é›†ï¼Œeval_datasetsæ˜¯ä¸€ä¸ªå­—å…¸ï¼Œkeyæ˜¯æ•°æ®é›†åç§°ï¼Œvalueæ˜¯æ•°æ®é›†æœ¬èº«
for ds_name, dataset in eval_datasets.items():
    # ç»Ÿè®¡å½“å‰æ•°æ®é›†çš„æ ·æœ¬æ•°é‡
    num_samples = len(dataset)
    print(f'\n{ds_name}: {num_samples} samples')

    # å°†æ¯æ¡æ ·æœ¬çš„ prompt è½¬æ¢ä¸ºæ¨¡å‹å¯ç”¨çš„è¾“å…¥æ ¼å¼
    # tokenizer.apply_chat_template å¯ä»¥å°†æ™®é€šæ–‡æœ¬åŒ…è£…æˆå¯¹è¯å¼è¾“å…¥ï¼Œå¹¶å¯é€‰æ‹©æ·»åŠ ç”Ÿæˆæç¤º
    test_texts = [
        tokenizer.apply_chat_template(sample['prompt'], add_generation_prompt=True, tokenize=False)
        for sample in dataset
    ]

    # ä½¿ç”¨ LoRA æƒé‡ç”Ÿæˆæ¨¡å‹è¾“å‡º
    outputs_with_lora = model.fast_generate(
        test_texts,
        sampling_params=sampling_params,
        lora_request=model.load_lora(f'./output/{model_name}_grpo'),  # åŠ è½½ LoRA æƒé‡
    )

    # ä¸ä½¿ç”¨ LoRA æƒé‡ç”Ÿæˆæ¨¡å‹è¾“å‡ºï¼ˆåŸºå‡†ï¼‰
    outputs_without_lora = model.fast_generate(
        test_texts,
        sampling_params=sampling_params,
        lora_request=None,
    )

    # åˆå§‹åŒ–ç»Ÿè®¡è®¡æ•°å™¨
    no_lora_format_cnt = lora_format_cnt = 0     # æ­£ç¡®æ ¼å¼æ•°é‡
    no_lora_answer_cnt = lora_answer_cnt = 0     # æ­£ç¡®ç­”æ¡ˆæ•°é‡
    no_lora_all_cnt = lora_all_cnt = 0           # æ ¼å¼å’Œç­”æ¡ˆéƒ½æ­£ç¡®çš„æ•°é‡

    # éå†æ¯æ¡æ ·æœ¬çš„è¾“å‡ºï¼Œåˆ†åˆ«å¯¹æ¯”ä½¿ç”¨å’Œä¸ä½¿ç”¨ LoRA çš„ç»“æœ
    for output_with_lora, output_without_lora, gt_answer in zip(outputs_with_lora, outputs_without_lora, dataset['answer']):

        # ================= ä½¿ç”¨ LoRA çš„ç»“æœ =================
        response_lora = output_with_lora.outputs[0].text  # è·å–ç”Ÿæˆçš„æ–‡æœ¬
        correct_format_lora = match_format.search(response_lora) is not None  # åˆ¤æ–­æ ¼å¼æ˜¯å¦æ­£ç¡®
        # å¦‚æœæ ¼å¼æ­£ç¡®ï¼Œåˆ™æå–ç­”æ¡ˆå­—æ¯ï¼Œå¦åˆ™ä¸º None
        extracted_guess_lora = match_answer_letter.search(response_lora).group(1) if correct_format_lora else None
        correct_answer_lora = extracted_guess_lora == gt_answer  # åˆ¤æ–­ç­”æ¡ˆæ˜¯å¦æ­£ç¡®
        correct_all_lora = correct_format_lora and correct_answer_lora  # æ ¼å¼å’Œç­”æ¡ˆéƒ½æ­£ç¡®

        # æ ¹æ®æ¡ä»¶ç´¯åŠ è®¡æ•°å™¨
        if correct_format_lora: lora_format_cnt += 1
        if correct_answer_lora: lora_answer_cnt += 1
        if correct_all_lora: lora_all_cnt += 1

        # ================= ä¸ä½¿ç”¨ LoRA çš„ç»“æœ =================
        response_no_lora = output_without_lora.outputs[0].text
        correct_format_no_lora = match_format.search(response_no_lora) is not None
        extracted_guess_no_lora = match_answer_letter.search(response_no_lora).group(1) if correct_format_no_lora else None
        correct_answer_no_lora = extracted_guess_no_lora == gt_answer
        correct_all_no_lora = correct_format_no_lora and correct_answer_no_lora

        if correct_format_no_lora: no_lora_format_cnt += 1
        if correct_answer_no_lora: no_lora_answer_cnt += 1
        if correct_all_no_lora: no_lora_all_cnt += 1

    # å°†ç»Ÿè®¡ç»“æœæ•´ç†æˆå­—å…¸ï¼Œæ–¹ä¾¿å±•ç¤º
    results[ds_name] = {
        'Without LoRA': {
            'Correct Format': f'{no_lora_format_cnt}/{num_samples} ({no_lora_format_cnt / num_samples * 100:.2f}%)',
            'Correct Answer': f'{no_lora_answer_cnt}/{num_samples} ({no_lora_answer_cnt / num_samples * 100:.2f}%)',
            'Correct Both': f'{no_lora_all_cnt}/{num_samples} ({no_lora_all_cnt / num_samples * 100:.2f}%)',
        },
        'With LoRA': {
            'Correct Format': f'{lora_format_cnt}/{num_samples} ({lora_format_cnt / num_samples * 100:.2f}%)',
            'Correct Answer': f'{lora_answer_cnt}/{num_samples} ({lora_answer_cnt / num_samples * 100:.2f}%)',
            'Correct Both': f'{lora_all_cnt}/{num_samples} ({lora_all_cnt / num_samples * 100:.2f}%)',
        },
        'Improvement': {  # è®¡ç®— LoRA ç›¸æ¯”ä¸ä½¿ç”¨ LoRA çš„æå‡
            'Correct Format': f'+{lora_format_cnt - no_lora_format_cnt} ({(lora_format_cnt - no_lora_format_cnt) / num_samples * 100:.2f}%)',
            'Correct Answer': f'+{lora_answer_cnt - no_lora_answer_cnt} ({(lora_answer_cnt - no_lora_answer_cnt) / num_samples * 100:.2f}%)',
            'Correct Both': f'+{lora_all_cnt - no_lora_all_cnt} ({(lora_all_cnt - no_lora_all_cnt) / num_samples * 100:.2f}%)',
        }
    }

    # ä½¿ç”¨ pandas DataFrame ç¾è§‚å±•ç¤ºç»“æœ
    display(pd.DataFrame(results[ds_name]).T)
```

# Inference

```python
# The FastLanguageModel.from_pretrained should be only called once. Otherwise, it will be OOM
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=f'./output/{model_name}_grpo', # Reload LoRA weights
    max_seq_length=max_seq_length,
    load_in_4bit=True,         # False for LoRA 16bit
    fast_inference=True,        # Enable vLLM fast inference
    max_lora_rank=lora_rank,
    gpu_memory_utilization=0.8, # Reduce if out of memory
)
FastLanguageModel.for_inference(model)
```

```python
def generate_with_reasoning(questions, max_completion_length=1024, system_prompt=SYSTEM_PROMPT):
    # Step 1: å°†æ¯ä¸ªé—®é¢˜åŒ…è£…æˆå¯¹è¯æ ¼å¼ï¼Œä¾¿äºæ¨¡å‹ç†è§£ä¸Šä¸‹æ–‡
    conversations = [[
        {'role': 'system', 'content': system_prompt},  # ç³»ç»Ÿè§’è‰²æç¤ºï¼Œç”¨äºæŒ‡å¯¼æ¨¡å‹è¡Œä¸º
        {'role': 'user', 'content': question},         # ç”¨æˆ·æé—®
    ] for question in questions]

    # Step 2: å°†å¯¹è¯æ ¼å¼è½¬æ¢ä¸ºæ¨¡å‹å¯æ¥å—çš„å­—ç¬¦ä¸²è¾“å…¥ï¼ˆä¸åštokenizeï¼‰
    prompts = [tokenizer.apply_chat_template(
        conversation,
        add_generation_prompt=True,  # åœ¨ç”¨æˆ·è¾“å…¥åæ·»åŠ  assistant æç¤ºç¬¦
        tokenize=False,              # è¿”å›å­—ç¬¦ä¸²è€Œé token
    ) for conversation in conversations]

    # Step 3: ä½¿ç”¨ tokenizer å°†å­—ç¬¦ä¸²æ‰¹é‡ç¼–ç ä¸ºæ¨¡å‹è¾“å…¥å¼ é‡
    inputs = tokenizer(prompts, return_tensors='pt', padding=True).to(model.device)

    # Step 4: è®°å½•ç”Ÿæˆå¼€å§‹æ—¶é—´
    start_time = time.time()

    with torch.no_grad():  # ä¸è®¡ç®—æ¢¯åº¦ï¼ŒèŠ‚çœæ˜¾å­˜
        output_ids = model.generate(
            **inputs,                   # è¾“å…¥å¼ é‡
            max_new_tokens=max_completion_length,  # ç”Ÿæˆæœ€å¤§é•¿åº¦
            temperature=0.6,             # æ§åˆ¶ç”Ÿæˆå¤šæ ·æ€§ï¼Œè¶Šä½è¶Šä¿å®ˆ
            top_p=0.95,                  # æ ¸é‡‡æ ·ï¼ˆnucleus samplingï¼‰ï¼Œé€‰æ‹©ç´¯è®¡æ¦‚ç‡å‰95%çš„è¯
            top_k=20,                     # top-ké‡‡æ ·ï¼Œé™åˆ¶é€‰æ‹©æœ€å¯èƒ½çš„20ä¸ªè¯
            do_sample=True,              # å¯ç”¨é‡‡æ ·è€Œéè´ªå¿ƒç”Ÿæˆ
            pad_token_id=tokenizer.eos_token_id,   # å¡«å……æ ‡è®°è®¾ç½®ä¸º eos
            repetition_penalty=1.1,      # æƒ©ç½šé‡å¤ï¼Œå‡å°‘é‡å¤æ¨ç†æ­¥éª¤
            length_penalty=1.0,          # é•¿åº¦æƒ©ç½šï¼Œ1.0è¡¨ç¤ºä¸­æ€§
            early_stopping=True,         # é‡åˆ°ç»ˆæ­¢ç¬¦æå‰åœæ­¢
            streamer=TextStreamer(tokenizer, skip_prompt=True),  # å®æ—¶æµå¼è¾“å‡ºï¼Œä¸æ˜¾ç¤ºæç¤ºéƒ¨åˆ†
        )

    # Step 5: è®°å½•ç”Ÿæˆç»“æŸæ—¶é—´
    end_time = time.time()
    inference_duration = end_time - start_time  # è®¡ç®—æ¨ç†è€—æ—¶
    num_generated_tokens = output_ids.shape[1] - inputs['input_ids'].shape[1]  # ç”Ÿæˆ token æ•°é‡

    # Step 6: æå–æ¨¡å‹ç”Ÿæˆçš„éƒ¨åˆ†ï¼ˆå»æ‰è¾“å…¥éƒ¨åˆ†ï¼‰
    output_ids = output_ids[:, inputs['input_ids'][0].shape[-1]:output_ids.shape[-1]]

    # Step 7: æ‰¹é‡è§£ç ç”Ÿæˆçš„ token ä¸ºå­—ç¬¦ä¸²ï¼Œè·³è¿‡ç‰¹æ®Š token
    responses = tokenizer.batch_decode(output_ids, skip_special_tokens=True)

    # Step 8: è¿”å›ç”Ÿæˆç»“æœã€è€—æ—¶å’Œç”Ÿæˆ token æ•°
    return responses, inference_duration, num_generated_tokens
```

```python
# ä»è¯„ä¼°æ•°æ®é›†å­—å…¸ä¸­å–å‡º MedQA_USLME çš„æµ‹è¯•é›†
test_dataset = eval_datasets['MedQA_USLME_test']

# è·å–æµ‹è¯•é›†ä¸­çš„ç¬¬3æ¡æ ·æœ¬çš„ç”¨æˆ·é—®é¢˜
# 'prompt' æ˜¯ä¸€ä¸ªå¯¹è¯åˆ—è¡¨ï¼Œ[-1]['content'] è¡¨ç¤ºå–æœ€åä¸€ä¸ªå¯¹è¯çš„æ–‡æœ¬å†…å®¹
medical_question = test_dataset[2]['prompt'][-1]['content']

# è·å–è¯¥æ ·æœ¬çš„æ ‡å‡†ç­”æ¡ˆï¼Œç”¨äºåç»­éªŒè¯æ¨¡å‹è¾“å‡º
expected_answer = test_dataset[2]['answer']

# æ‰“å°é—®é¢˜æ–‡æœ¬ï¼Œä¾¿äºè§‚å¯Ÿ
print(medical_question, '\n\n===== Response =====')

# è°ƒç”¨è‡ªå®šä¹‰å‡½æ•° generate_with_reasoning è¿›è¡Œæ¨ç†ç”Ÿæˆ
# è¾“å…¥æ˜¯ä¸€ä¸ªåŒ…å«å•ä¸ªé—®é¢˜çš„åˆ—è¡¨ï¼Œmax_completion_length æ§åˆ¶ç”Ÿæˆçš„æœ€å¤§é•¿åº¦
medical_responses, inference_duration, num_generated_tokens = generate_with_reasoning(
    [medical_question],
    max_completion_length
)

# å› ä¸ºæˆ‘ä»¬åªè¾“å…¥äº†ä¸€ä¸ªé—®é¢˜ï¼Œç”Ÿæˆç»“æœæ˜¯åˆ—è¡¨ï¼Œè¿™é‡Œå–ç¬¬ä¸€ä¸ªå…ƒç´ 
medical_response = medical_responses[0]

# æ‰“å°æ¨ç†è€—æ—¶ï¼ˆç§’ï¼‰
print('Inference time (secs):', inference_duration)

# æ‰“å°æ¨¡å‹ç”Ÿæˆçš„ token æ•°é‡ï¼Œç”¨äºåˆ†æç”Ÿæˆé•¿åº¦å’Œæ€§èƒ½
print('Generated tokens:', num_generated_tokens)
```
